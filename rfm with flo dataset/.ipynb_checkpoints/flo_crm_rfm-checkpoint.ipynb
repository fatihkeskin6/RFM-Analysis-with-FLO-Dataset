```python
import pandas as pd
from datetime import datetime as dt
```

```python
df_ = pd.read_csv('./datasets/flo_data_20k.csv')
```

```python
df = df_.copy()
```

```python
################## PREPROCESSING $$$$$$$$$$$$$$$$$$
```

```python
# Print the dataset from here after every single action
df.head()
```

```python
# Floats to integer
df['order_num_total_ever_online'] = df['order_num_total_ever_online'].astype(int)
df['order_num_total_ever_offline'] = df['order_num_total_ever_offline'].astype(int)
```

```python
# Strings to Date
df['first_order_date'] = pd.to_datetime(df['first_order_date'])
df['last_order_date'] = pd.to_datetime(df['last_order_date'])
df['last_order_date_online'] = pd.to_datetime(df['last_order_date_online'])
df['last_order_date_offline'] = pd.to_datetime(df['last_order_date_offline'])
```

```python
# Check for null values
df.isnull().sum()
```

```python
# Check the statistics of values of dataframe [Outliers]
df.describe().T
```

```python
# Sum of both online and offline purchases of each customer and number of purchases at both platforms included.
df['total_number_of_order'] = df['order_num_total_ever_online']+ df['order_num_total_ever_offline']
df['sum_of_customer_values'] = df['customer_value_total_ever_offline']+df['customer_value_total_ever_online']
```

```python
# Total customer value and sum of the orders of each platform,
df.groupby('order_channel')[['total_number_of_order','sum_of_customer_values']].sum()
```

```python
# List top 10 customer which have the most order 
df.sort_values(by=['total_number_of_order'],ascending=False).head(n=10)
```

```python
############# Calculations of RFM Metrics $$$$$$$$$$$$$$
```

```python
# Creating current date variable to get recency value (It is  2 days after the last order)
df['last_order_date'].max()

# two days after code above
today_date= pd.to_datetime('2021-06-01')
today_date
```

```python
# Creation of RFM Metrics
rfm = df.groupby('master_id').agg({
                            'last_order_date':lambda last_order_date:(today_date-last_order_date.max()).days,
                            'total_number_of_order':lambda total_number_of_order:total_number_of_order.sum(),
                            'sum_of_customer_values':lambda sum_of_customer_values: sum_of_customer_values.sum()
})

rfm = rfm.rename(columns={'last_order_date':'Recency','total_number_of_order':'Frequency','sum_of_customer_values':'Monetary'})
rfm
```

```python
############ Creation of RFM Scores and Total RF Score based on RFM Metrics $$$$$$$$$$$$$
rfm['recency_score'] = pd.qcut(rfm['Recency'],5,labels=[5,4,3,2,1])
rfm['frequency_score']= pd.qcut(rfm['Frequency'].rank(method='first'),5,labels=[1,2,3,4,5])
rfm['monetary _score']= pd.qcut(rfm['Monetary'],5,labels=[1,2,3,4,5])
rfm['RF Score'] =(rfm['recency_score'].astype(str) +
                                 rfm['frequency_score'].astype(str))
rfm
```

```python
######### Segmentation based on RF Scores $$$$$$$$$$
```

```python
seg_map = {
        r'[1-2][1-2]': 'hibernating',
        r'[1-2][3-4]': 'at_risk',
        r'[1-2]5': 'cant_loose',
        r'3[1-2]': 'about_to_sleep',
        r'33': 'need_attention',
        r'[3-4][4-5]': 'loyal_customers',
        r'41': 'promising',
        r'51': 'new_customers',
        r'[4-5][2-3]': 'potential_loyalists',
        r'5[4-5]': 'champions'
    }
```

```python
rfm['Segment'] = rfm['RF Score'].replace(seg_map,regex=True)
```

```python
rfm
```

```python
#Count of customers at each segment
rfm.groupby('Segment').size()
```

```python
######### CASE : List all the customers which has segment of 'champions' OR 'loyal_customers' 
#                                                                         AND has order from WOMEN category
#                and extract it as csv.
```

```python
champs_and_loyals=rfm[rfm['Segment'].isin(['champions','loyal_customers'])]
```

```python
champs_and_loyals=champs_and_loyals.reset_index(inplace=False)
champs_and_loyals_ids= champs_and_loyals['master_id']
champs_and_loyals
```

```python
# Customers which have the master_id on dataset of 'champs_and_loyals' AND have women in interested categories
csv_file=df.loc[df['master_id'].isin(champs_and_loyals_ids)&df['interested_in_categories_12'].str.contains('KADIN')]
```

```python
# Extract the dataframe as csv file
csv_file.to_csv('champions_loyals_category_women.csv')
```
